```python
#Changing Value for Male Female String to numeric values , male=1 and female=2
def getNumber(str):
    if str=="male":
        return 1
    else:
        return 2
titanic["gender"]=titanic["Sex"].apply(getNumber)
titanic.head()
```

```python
del titanic["Sex"]
titanic.head()
```

```python
#Pie Chart
males = (titanic['gender'] == 1).sum()
females = (titanic['gender'] == 2).sum()

p = [males, females]
plt.pie(p,
       labels = ['Male', 'Female'],
       colors = ['blue', 'red'],
       explode = (0.15, 0),
       startangle = 0)
plt.axis('equal')
plt.show()
```
